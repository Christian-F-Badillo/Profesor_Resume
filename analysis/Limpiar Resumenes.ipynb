```python
import pandas as pd
import re
```

```python
URL = 'https://raw.githubusercontent.com/Christian-F-Badillo/Profesor_Resume/refs/heads/master/data/reviews_plus_score.csv'
data = pd.read_csv(URL)
```

```python
data.head(10)
```

```python
merge_reviews = data.groupby("profesor")['comentario'].apply(lambda x: ' '.join(x)).reset_index()
```

```python
merge_reviews.head()
```

```python
URL2 = 'https://raw.githubusercontent.com/Christian-F-Badillo/Profesor_Resume/refs/heads/master/data/summaries.txt'
summary_data = pd.read_csv(URL2, sep="\t", header=None)
```

```python
summary_data.head()
```

```python
summary_data.columns = ['resumen']
summary_data.head()
```

```python
# Eliminamos el número de incio de cada resumen ejemp: 1. blablabla
summary_data['resumen'] = summary_data['resumen'].apply(lambda x: re.sub(r'^\d+\.\s*', '', x))
```

```python
summary_data.head()
```

```python
# Combinamos los comentarios con los resúmenes
merged_data = pd.merge(merge_reviews, summary_data, left_index=True, right_index=True)
```

```python
merged_data.head(10)
```

```python
print(merged_data.iloc[488, [0, 2]].tolist())
```

```python
# Quitamos los comentarios y solo dejamos los resúmenes
merged_data = merged_data.drop(columns=['comentario'])
```

```python
# Arreglamos los nombres de los profesores a capitalizar la primera letra de cada palabra
merged_data['profesor'] = merged_data['profesor'].apply(lambda x: x.title())
merged_data.head(10)
```

```python
# Guardamos el DataFrame resultante
merged_data.to_csv('summaries.csv', index=False)
```
