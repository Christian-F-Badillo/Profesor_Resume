```python
import pandas as pd
```

```python
data_scores = pd.read_csv('https://raw.githubusercontent.com/Christian-F-Badillo/Profesor_Resume/refs/heads/master/data/reviews_plus_score.csv')
data_reviews = pd.read_csv('https://raw.githubusercontent.com/Christian-F-Badillo/Profesor_Resume/refs/heads/master/data/summaries.csv')
data_tags = pd.read_csv('https://raw.githubusercontent.com/Christian-F-Badillo/Profesor_Resume/refs/heads/master/data/merged_tags.csv')
data_rating = pd.read_csv('https://raw.githubusercontent.com/Christian-F-Badillo/Profesor_Resume/refs/heads/master/data/tabla_profesores.csv')
data_enlaces_origin = pd.read_csv('https://raw.githubusercontent.com/Christian-F-Badillo/Profesor_Resume/refs/heads/master/data/enlaces_profesores_clean.csv')
```

```python
data_scores.head()
```

```python
scores = data_scores.drop(columns=['comentario', 'review']).copy()
scores.head()
```

```python
# Stars Dict
map_stars = {
    "1 star": 1,
    "2 stars": 2,
    "3 stars": 3,
    "4 stars": 4,
    "5 stars": 5
}

scores['rating'] = scores['rating'].map(map_stars)
scores.head()
```

```python
# Capitalizamos los nombres de los profesores
scores['profesor'] = scores['profesor'].str.title()
scores.head()
```

```python
# Pasamos a Wide format nuestros scores, poniendo la información en forma de cadena de texto separada por comas
scores_wide_rating = scores.groupby('profesor')['rating'].apply(lambda x: ', '.join(map(str, x))).reset_index()
scores_wide_date = scores.groupby('profesor')['fecha'].apply(lambda x: ', '.join(map(str, x))).reset_index()
scores_wide_calidad = scores.groupby('profesor')['calidad_general'].apply(lambda x: ', '.join(map(str, x))).reset_index()
scores_wide_facilidad = scores.groupby('profesor')['facilidad'].apply(lambda x: ', '.join(map(str, x))).reset_index()
```

```python
scores_wide = pd.merge(scores_wide_rating, scores_wide_date, on='profesor')
scores_wide = pd.merge(scores_wide, scores_wide_calidad, on='profesor')
scores_wide = pd.merge(scores_wide, scores_wide_facilidad, on='profesor')
scores_wide.head()
```

```python
# Combinamos con tags pero primero capitalizamos los nombres de los profesores
data_tags['profesor'] = data_tags['profesor'].str.title()
scores_wide = pd.merge(scores_wide, data_tags, on='profesor', how='left')
scores_wide.head()
```

```python
data_rating.drop(columns=['dep'], inplace=True)
data_rating.head()
```

```python
data_rating.columns = ['profesor', 'num_reviews', 'promedio']
data_rating['profesor'] = data_rating['profesor'].str.title()
data_rating.head()
```

```python
scores_wide = pd.merge(scores_wide, data_rating, on='profesor', how='left')
```

```python
scores_wide.head()
```

```python
# Combinamos con los resumenes de los profesores
data_reviews['profesor'] = data_reviews['profesor'].str.title()
scores_wide = pd.merge(scores_wide, data_reviews, on='profesor', how ='left')
```

```python
scores_wide.head()
```

```python
# Combinamos con los enlaces de los profesores
data_enlaces_origin['name'] = data_enlaces_origin['name'].str.title()
data_enlaces_origin.columns = ['enlace', 'profesor']
data_enlaces_origin.head()
```

```python
# Hacemos el merge final con los enlaces
scores_wide = pd.merge(scores_wide, data_enlaces_origin, on='profesor', how='left')
```

```python
# Guardamos todo en un CSV
scores_wide.to_csv('professor_data.csv', index=False)
```
